```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn import linear_model
from scipy import stats
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_excel("D:/New folder/Semester 3/ADY301m/file csv/df_data_house.xlsx")
data.head()
```

```python
x=data[["bedrooms", "bathrooms", "sqft_living", "sqft_lot","waterfront","floors", "grade"]]
y=data["price"]
y_bi=data["Binary_Price"]

x=sm.add_constant(x)

model=sm.OLS(y,x).fit()

print(model.summary())
```

```python
# Há»“i quy logistic
model2=sm.Logit(y_bi, x).fit()

print(model2.summary())
```

```python
residuals = model.resid

# TÃ­nh MSE
MSE = np.mean(residuals**2)
print("MSE:", MSE)

# TÃ­nh RMSE
RMSE = np.sqrt(MSE)
print("RMSE:", RMSE)
```
