```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_csv('D:/New folder/Semester 3/ADY301m/file csv/Temp_and_rain.csv')
```

```python
print(data.head())
print(data.info())

```

```python
#check null value
missing_value=data.isnull().sum()
print(missing_value)
#handle missing value
data.fillna(data.mean(),inplace=True)
```

```python
#tính lượng mưa trung bình và nhiêt độ các năm
average_rain=data.groupby('Year')['rain'].mean()
average_temp=data.groupby("Year")["tem"].mean()
print(average_rain, average_temp)
```

```python
#vẽ biểu đồ trung bình lương mưa qua các năm
plt.figure(figsize=(12, 6))
plt.plot(average_rain.index, average_rain.values, color="blue", label="rain")
plt.title("Average rainfall(1901-2015)")
plt.xlabel("Year")
plt.ylabel("Rain")
plt.legend()
plt.show()
```

```python
#vẽ biểu đồ trung bình nhiet do qua các năm
plt.figure(figsize=(12, 6))
plt.plot(average_temp.index, average_temp.values, color="blue", label="temperature")
plt.title("Average temperature(1901-2015)")
plt.xlabel("Year")
plt.ylabel("temperature")
plt.legend()
plt.show()
```
