```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
print(sns.get_dataset_names())
```

```python
sns.set_theme()
```

```python
tips_df= sns.load_dataset("tips")
tips_df.head()
```

```python
sns.histplot(tips_df["total_bill"]);
```

```python
sns.kdeplot(tips_df["total_bill"]);
```

```python
sns.displot(tips_df, x="total_bill", col="time", kde=True)
```

# Barplot


```python
sns.barplot(tips_df, x="sex", y="tip", estimator=np.mean, color="grey");
```

```python
tips_df["sex"].value_counts()
sns.countplot(tips_df, x="sex");
```

# BOXPLOT


```python
sns.boxplot(tips_df, x="day", y="total_bill",hue="sex", palette="Blues");
plt.legend(loc=0);
```

# FACEGRID



```python
new_fg=sns.FacetGrid(tips_df, col="sex",hue="smoker" , 
                     col_order=["Female", "Male"],
                     palette="Set2",
                     height=4, aspect=1.3)
new_fg.map(sns.scatterplot, "total_bill", "tip", s=100,alpha=.8);
new_fg.add_legend();
```

```python
penguins_df=sns.load_dataset('penguins')
penguins_df.head()
```

```python
sns.jointplot(penguins_df, x="flipper_length_mm", y="bill_length_mm", hue="species")
```

#Pair Plot


```python
sns.pairplot(penguins_df, hue="species")
```

# HEAD MAPS

```python
flight_df=sns.load_dataset("flights")
flight_df.head()
```

```python
flights=pd.pivot_table(flight_df, index="month",columns="year" ,values="passengers"  );
flights
```

```python
sns.heatmap(flights, cmap="Blues",linecolor="white",linewidths=1);
```
