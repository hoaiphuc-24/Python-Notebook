```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
data=pd.read_csv('D:/New folder/Semester 3/ADY301m/file csv/Healthcare-Diabetes.csv')
print(data.head())
print(data.shape)
```

```python
#check missing value
print(data.isnull().sum())
#data.fillna(data.mean(),inplace=True)
#Basic Statistics
print(data.describe())
```

```python
#Visualization
fig,((a1,a2) ,(a3,a4))=plt.subplots(nrows=2,ncols=2, figsize=(12,8))
#a1 là bd của cot Pregnency
bin1 = [0, 3, 5, 7, 10, 17]
label1 = ['1-3', '3-5', '5-7', '7-10', '10-17']
data['Pregnancy Group'] = pd.cut(data['Pregnancies'], bins=bin1, labels=label1, right=False)
pregnancy_group_counts = data['Pregnancy Group'].value_counts(sort=False)
pregnancy_group_counts.plot(kind='bar', ax=a1)
a1.set_xlabel('Pregnancy Groups')
a1.set_ylabel('Count')
a1.set_title('Distribution of Pregnancies by Groups')
a1.set_xticklabels(a1.get_xticklabels(), rotation=45, ha='right')

#a2 là bd của cột Glucoso
data["Glucose Group"] = pd.cut(data["Glucose"], bins=[30,50,70,80,110,150,170,200])
glucose_counts = data['Glucose Group'].value_counts(sort=False)
glucose_counts.plot(kind="bar", ax=a2)
a2.set(title="Glucose", xlabel= "Glucose Group", ylabel="Glucose")
a2.set_xticklabels(a2.get_xticklabels(), rotation=45, ha='right')

#a3
a3.hist(x=data["BloodPressure"], bins=6, color='lightblue', edgecolor='black')
a3.set(title="Blood Pressure Distribution", xlabel="Blood Pressure", ylabel="Frequency")
#a4
a4.hist(data["SkinThickness"], bins=[0,10,20,30,40,50,60], edgecolor="blue")
a4.set(title="SkinThickness")
plt.tight_layout()
plt.show()
```
