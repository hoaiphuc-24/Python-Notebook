```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from selenium import webdriver 
import time 
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("D:/New folder/Semester 3/ADY301m/file csv/aus_real_estate.csv")

df.head()
```

```python
st=df["Price"].describe()
print(f"Thống kê mô tả:")
st
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=25, kde=True, color='skyblue')
plt.title('Distribution of Property Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
avg_price_city = df.groupby('City')['Price'].mean().sort_values(ascending=False)
sns.barplot(x=avg_price_city.index, y=avg_price_city.values, palette='viridis')
plt.title('Average Price by City')
plt.xlabel('City')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show();
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Type', y='Price', data=df, palette='coolwarm')
plt.title('Price Comparison by Property Type')
plt.xlabel('Property Type')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
avg_garage_city = df.groupby('City')['Garage'].sum().sort_values(ascending=False)
sns.barplot(x=avg_garage_city.index, y=avg_garage_city.values)
plt.title('Average garage by City')
plt.xlabel('City')
plt.ylabel('Average garage')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show();
```
